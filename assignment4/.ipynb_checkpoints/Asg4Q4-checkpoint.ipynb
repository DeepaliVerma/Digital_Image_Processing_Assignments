```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import copy
```

```python
def verify(inp, s, flag):
    if flag:
        tmp = inp==s
        return (s==-1).sum() + (tmp==1).sum() == 9
    for i in range(3):
        for j in range(3):
            if s[i,j] == -1:
                continue
            elif s[i,j] != inp[i,j]:
                return 0
    return 1
    

def apply_filter(img, s, flag=0):
    h, w = img.shape
    copy = img.copy()
    for i in range(1,h-1):
        for j in range(1,w-1):
            mat = img[i-1:i+2, j-1: j+2]
            if verify(mat, s, flag) == 1:
                copy[i,j] = 0
    return copy
```

```python
def unite(a, linked):
    if linked[a] == a:
        return a
    else:
        linked[a] = unite(linked[a], linked)
        return linked[a]
def unionSet(a, b, linked):
    linked[unite(a, linked)] = unite(b, linked)
def connComponents(im):
    next_label = 1;
    labels = np.zeros(im.shape, dtype=int)
    linked = [0]
    h, w = im.shape
    for row in range(h):
        for column in range(w):
            if im[row, column] != 0:
                neighbors = []
                if row > 0:
                    if im[row-1][column] == im[row, column]:
                        neighbors.append((row-1, column))
                if row > 0 and column + 1 < im.shape[1]:
                    if im[row-1, column+1] == im[row, column]:
                        neighbors.append((row-1, column+1))
                if row > 0 and column > 0:
                    if im[row-1, column-1] == im[row, column]:
                        neighbors.append((row-1, column-1))
                if column > 0:
                    if (im[row, column-1] == im[row, column]):
                        neighbors.append((row, column-1))

                if len(neighbors) == 0:
                    linked.append(next_label)
                    labels[row, column] = next_label
                    next_label += 1
                
                else:
                    L = [labels[r, c] for r, c in neighbors]
                    labels[row, column] = min(L)
                    for label in L:
                        unionSet(label, labels[row, column], linked)
                        
    for row in range(im.shape[0]):
        for column in range(im.shape[1]):
            if im[row, column] != 0:
                labels[row, column] = unite(labels[row, column], linked)
    return labels
```

```python
im = cv2.imread('./images/objects.png',0);
bin_im = np.uint8(im > 128)
plt.imshow(bin_im,cmap = cm.gray)
```

### Part 1

```python
neg_bin_im = connComponents(1 - bin_im)
neg_bin_im[neg_bin_im == 1] = 0
neg_bin_im[neg_bin_im !=0 ] = 1
plt.imshow(neg_bin_im,cmap=cm.gray)
holes = neg_bin_im
```

```python
restored = bin_im + holes
plt.imshow(restored,cmap=cm.gray)
```

```python
original = connComponents(bin_im)
filled = copy.deepcopy(restored)
restored = connComponents(restored)
plt.figure(figsize=(15, 15))
plt.subplot(1, 2, 1)
plt.imshow(original, cmap = cm.gist_stern)
plt.title('Original')
plt.subplot(1, 2, 2)
plt.imshow(restored, cmap= cm.gist_stern)
plt.title('Restored')
plt.show()
```

```python
restored_unique, restored_count = np.unique(restored, return_counts=True)
og_unique, og_count = np.unique(original, return_counts=True)
holed_objects_bool = restored_count - og_count > 0
print("Number of objects with holes = ",np.sum(holed_objects_bool))
```

```python
only_holed = copy.deepcopy(original)
for i in range(holed_objects_bool.shape[0]):
    if holed_objects_bool[i] == False:
        only_holed[only_holed == og_unique[i]] = 0
        
plt.figure(figsize=(15, 15))
plt.subplot(1, 2, 1)
plt.imshow(original, cmap = cm.gist_stern)
plt.title('Original')
plt.subplot(1, 2, 2)
plt.imshow(only_holed, cmap= cm.gist_stern)
plt.title('Only holed objects')
plt.show() 
        
```

### Part 2

```python
kernel = np.array([[1, 1, 1], [1,-1, -1], [1, -1, -1]])
only_squares = apply_filter(filled,kernel)
for i in range(40):
    only_squares = apply_filter(only_squares, kernel)

plt.figure(figsize=(7, 7))
plt.imshow(only_squares,cmap=cm.gray)    
```

```python
inv = connComponents(1-only_squares)
inv[inv==1] = 0
plt.imshow(inv,cmap = cm.nipy_spectral)
print("Number of squares = ",np.unique(inv).shape[0]-1)
```

### Part 3

```python
inv[inv!=0] = 1
filled_squares = inv
plt.imshow(filled_squares,cmap = cm.gray)
```

```python
only_holed_filled = copy.deepcopy(restored)
for i in range(holed_objects_bool.shape[0]):
    if holed_objects_bool[i] == False:
        only_holed_filled[only_holed_filled == restored_unique[i]] = 0
only_holed_filled[only_holed_filled!=0] = 1
plt.imshow(only_holed_filled, cmap= cm.gray)
```

```python
only_holed_squares = copy.deepcopy(bin_im)
only_holed_squares[filled_squares * only_holed_filled==0] = 0
plt.imshow(only_holed_squares,cmap = cm.gray)
```

### Part 4

```python
only_full = filled - only_holed_filled
only_circles = (only_full - filled_squares)
only_circles[only_circles<0]= 0
# plt.imshow(only_circles,cmap = cm.gray)
only_circles = only_circles.astype(np.uint8)
# np.unique(only_circles)
kernel = np.ones((2,2),np.uint8)
eroded = cv2.erode(only_circles,kernel)
plt.imshow(eroded,cmap = cm.gray)
plt.title("Circles")
```
